BFS implementation

```python
def bfs(G, start):
    visited = set([start])
    Q = deque()
    Q.append(start)

    while Q:
        u = Q.popleft()
        print(u, end=" ")

        for v in G.neighbors(u):
            if v not in visited:
                visited.add(v)
                Q.append(v)
```
